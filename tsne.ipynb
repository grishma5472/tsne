```python
from sklearn.manifold import TSNE
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris #loading iris dataset
import numpy as np
```

```python
#The Iris dataset consists of 150 samples of iris flowers
#from three different species: Setosa, Versicolor, and Virginica
#Each sample includes four features:
#sepal length, sepal width, petal length, and petal width.

# n(categories)=3 & n(variables)=4

iris= load_iris()
data=pd.DataFrame(iris.data,columns=iris.feature_names)
data["Species"]=iris.target
```

```python
iris_unique = data.drop_duplicates()

# Select features (petal length, petal width, sepal length, sepal width)
iris_matrix = iris_unique.iloc[:, :4].values

# Set a seed for reproducibility
np.random.seed(42)  # Import numpy as np if not already

# Perform t-SNE dimensionality reduction
tsne = TSNE(n_components=2)  # Adjust n_components for desired output dimension
tsne_out = tsne.fit_transform(iris_matrix)

# Create the scatter plot
plt.scatter(tsne_out[:, 0], tsne_out[:, 1], c=iris_unique["Species"])

# Add labels and title
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.title("t-SNE of Iris Dataset (Colored by Species)")

# Display the plot
plt.show()
```
